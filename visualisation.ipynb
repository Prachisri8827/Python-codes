```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = sns.load_dataset('iris')
df
```

```python
#bar plot for sepal length
_ = sns.barplot(x='species', y='sepal_length', data=df)
```

```python
sns.barplot(x='species', y='sepal_width', data=df)
```

```python
sns.barplot(x='species', y='petal_length', data=df)
```

```python
sns.barplot(x='species', y='petal_width', data=df)
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(10,10), nrows=2, ncols=2)
sns.barplot(df,x='species', y='sepal_length', ax=ax[0,0])
sns.barplot(df,x='species', y='sepal_width', ax=ax[0,1])
sns.barplot(df,x='species', y='petal_length', ax=ax[1,0])
sns.barplot(df,x='species', y='petal_width', ax=ax[1,1])
```
